```python
import numpy as np
from matplotlib import pyplot as plt
from linear_advection import Advection1D
```

```python
problem = Advection1D(c=1, domain=[-1,1], nx=129)
sol1 = problem.solve_backward_euler(dt=0.005, t_start=0.0, t_stop=2.0)
sol2 = problem.solve_ivp(dt=0.005, t_start=0.0, t_stop=2.0).y

plt.plot(problem.x, sol1[:,0], 'b:', label='initial condition u(x,0)')
plt.plot(problem.x, sol1[:,-1], 'r-', label='solution at final time u(x, 2) -- BE')
plt.plot(problem.x, sol2[:,-1], 'r:', label='solution at final time u(x, 2) -- scipy')
plt.xlabel('x')
plt.show()

problem.dx
```

```python
def gaussian(x):
    nx = x.shape[0]
    return np.exp(-x**2/4)

problem = Advection1D(c=1, domain=[-1,1], nx=129, x_init=gaussian)
sol1 = problem.solve_backward_euler(dt=0.005, t_start=0.0, t_stop=2.0)
sol2 = problem.solve_ivp(dt=0.005, t_start=0.0, t_stop=2.0).y

plt.plot(problem.x, sol1[:,0], 'b:', label='initial condition u(x,0)')
plt.plot(problem.x, sol1[:,-1], 'r-', label='solution at final time u(x, 2) -- BE')
plt.plot(problem.x, sol2[:,-1], 'r:', label='solution at final time u(x, 2) -- scipy')
plt.xlabel('x')
plt.show()
```
